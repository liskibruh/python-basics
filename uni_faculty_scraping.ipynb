```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

pd.set_option('display.max_colwidth', 100)
```

```python
faculties = ['cs', 'ee', 'sh']
urls = [f"http://pwr.nu.edu.pk/{each_faculty}-faculty/" for each_faculty in faculties]

headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; rv:91.0) Gecko/20100101 Firefox/91.0',
'Accept-Language': 'en-US,en;q=0.9'
}
```

```python
soups = []
for each_url in urls:
    response = requests.get(each_url, headers=headers)
    soups.append(bs(response.text))
```

```python
profile_sections = soups[0].find_all('div', class_='col-xs-12 col-sm-6 col-md-3 sm-text-center mb-30 mb-sm-30')
profile_sections.extend(soups[1].find_all('div', class_='col-xs-12 col-sm-6 col-md-3 sm-text-center mb-30 mb-sm-30'))
profile_sections.extend(soups[2].find_all('div', class_='col-xs-12 col-sm-6 col-md-3 sm-text-center mb-30 mb-sm-30'))
```

```python
employee_roles = []
employee_names = []
employee_images = []
employee_emails = []
employee_readmore = []
employee_department = []

for i,each_profile in enumerate(profile_sections):
    if(i<=21):
        employee_department.insert(i,"CS")
    elif(i>21 and i<=34):
        employee_department.insert(i,"EE")
    else:
        employee_department.insert(i,"SH")
    #grabbing names
    employee_names.append(each_profile.h4.text)
    #grabbing roles
    employee_roles.append(each_profile.h5.text)
    #grabbing images
    img_tag = each_profile.find('img')
    employee_images.append("http://pwr.nu.edu.pk/"+(img_tag.get('src')))
    #grabbing emails
    a_tags = each_profile.find_all('a')
    employee_emails.append((a_tags[1].get('href')).split(':')[1])
    #grabbing read more links
    employee_readmore.append("http://pwr.nu.edu.pk/"+a_tags[0].get('href'))
```

```python
data = {
    "Name" : employee_names,
    "Role" : employee_roles,
    "Email" : employee_emails,
    "Department": employee_department,
    "Image" : employee_images,
    "Read More": employee_readmore
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv("scraped_uni_faculty.csv")
```
